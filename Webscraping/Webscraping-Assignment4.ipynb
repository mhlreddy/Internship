1. Scrape the details of most viewed videos on YouTube from Wikipedia.Url 
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A)Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
# Initialize the Chrome driver
driver1 = webdriver.Chrome()
# Opening web page on automated Chrome browser
driver1.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')  
# Pausing the webdriver for 4 seconds
time.sleep(4)
```

```python
# Empty list to strore scraped rank data
rank = []

# Locating rank
rank_list = driver1.find_elements(By.XPATH,"//div[@id='mw-content-text']/div[1]/table[1]/tbody/tr") 
# Scraping the rank of most viewedYouTube video
for i in rank_list:
    try:
        if i.text is None:
            rank.append("--")
        else:
            rank.append(rank_list.index(i)+1)
    except NoSuchElementException as e:
        rank.append("--")
        
print(len(rank))
print(rank)       
```

```python
# Empty list to store name of the video
video_name = []

# scraping most viewed video Name and appending list 
vn = driver1.find_elements(By.XPATH,"//div[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[1]")
for i in vn:
    if i.text is None :
        video_name.append("--") 
    else:
        video_name.append(i.text)
        
print(len(video_name))
print(video_name)
```

```python
# Empty list to store artist of the video
artist_name = []

# Scraping the artist name
artist = driver1.find_elements(By.XPATH,"//div[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[2]")
for i in artist:
    if i.text is None :
        artist_name.append("--") 
    else:
        artist_name.append(i.text)
        
print(len(artist_name),artist_name)
```

```python
# Empty list to store uploaded date of the video
upload_date = []

#scraping the upload_date 
date = driver1.find_elements(By.XPATH,"//div[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[4]")
for i in date:
    if i.text is None :
        upload_date.append("--") 
    else:
        upload_date.append(i.text)
        
print(len(upload_date),upload_date)
```

```python
# Empty list to store no. of views of the video
views = []

#scraping views
vs = driver1.find_elements(By.XPATH,"//div[@id='mw-content-text']/div[1]/table[1]/tbody/tr/td[3]")
for i in vs:
    if i.text is None :
        views.append("--") 
    else:
        views.append(i.text)
        
print(len(views),views)
```

```python
# Creating adata frame
mvv_df = pd.DataFrame({'Rank': rank, 'Video title': video_name, 'Artist name': artist_name, 'Uplod date': upload_date, 'Views in billion': views})
mvv_df
```

```python
#Close the browser session 
driver1.close()
```

```python
#scraping time
tim = driver2.find_elements(By.XPATH,"//div[@class='match-date-info']/div[2]")
for i in tim:
    if i.text is None :
        time.append("--") 
    else:
        time.append(i.text)
        
print(len(time),time)
```

3. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/
You have to find following details: A) Rank
B) State
C) GSDP(18-19)- at current prices
D) GSDP(19-20)- at current prices
E) Share(18-19)
F) GDP($ billion)
Note:- From statisticstimes home page you have to reach to economy page through code.

```python
# Initialize the Chrome driver
driver3 = webdriver.Chrome()
# Opening web page on automated Chrome browser
driver3.get('http://statisticstimes.com/')  
```

```python
# Click the Economy tab
economy_page = driver3.find_element(By.XPATH,"/html/body/div[2]/div[1]/div[2]/div[2]/button")
economy_page.click()
# select India tab
ind_page = driver3.find_element(By.XPATH,"//div[@class='navbar']/div[2]/div/a[3]")
ind_page.click()
```

```python
rank = []
#scraping the Rank 
state_rank = driver3.find_elements(By.XPATH, "//td[@class='data1']")
for i in state_rank:
    if i.text is None :
        rank.append("--") 
        
    else:
        rank.append(i.text)
        
for elt in rank:
    if elt == ',' or elt == '':
        rank.remove(elt)
        
print(len(rank),rank)
```

```python
state_name = []

#scraping the State names 
states = driver3.find_elements(By.XPATH, "//td[@class='name']")
for i in states:
    if i.text is None :
        state_name.append("--") 
    else:
        state_name.append(i.text)
        
for elt in state_name:
    if elt == 'India':
        state_name.remove(elt)
        
print(len(state_name),state_name)
```

```python
GSDP_2223 = []
#scraping the GSDP of 2022 -23 
gsdp2223 = driver3.find_elements(By.XPATH, "//table[@class='display dataTable']/tbody/tr/td[4]")
for i in gsdp2223:
    if i.text is None :
        GSDP_2223.append("-") 
    else:
        GSDP_2223.append(i.text)
        
print(len(GSDP_2223),GSDP_2223)
```

```python
GSDP_2122 = []
#scraping the GSDP of 2021 -22 
gsdp2122 = driver3.find_elements(By.XPATH, "//table[@class='display dataTable']/tbody/tr/td[5]")
for i in gsdp2122:
    if i.text is None :
        GSDP_2122.append("-") 
    else:
        GSDP_2122.append(i.text)
        
print(len(GSDP_2122),GSDP_2122)
```

```python
share_2122 = []
#scraping share in 2021 -22 
share = driver3.find_elements(By.XPATH, "//table[@class='display dataTable']/tbody/tr/td[6]")
for i in share:
    if i.text is None :
        share_2122.append("-") 
    else:
        share_2122.append(i.text)
        
print(len(share_2122),share_2122)
```

```python
GDP_2021 = []
#scraping the GDP of 2021 -22 
gdp = driver3.find_elements(By.XPATH, "//table[@class='display dataTable']/tbody/tr/td[7]")
for i in gdp:
    if i.text is None :
        GDP_2021.append("-") 
    else:
        GDP_2021.append(i.text)
        
print(len(GDP_2021),GDP_2021)
```

```python
# Creating adata frame
states_df = pd.DataFrame({'Rank': rank, 'State': state_name, 'GSDP (2022-23)': GSDP_2223, 'GSDP (2021-22)': GSDP_2122, 'Share in GDP (2021-22)': share_2122, 'GDP ($ billion)': GDP_2021})

states_df
```

```python
#Close the browser session 
driver3.close()
```
